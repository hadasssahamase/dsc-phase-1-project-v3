Data Preparation
This involves data cleaning by finding and clearing missing values,dropping rows or columns,filtering data,adding relevant columns

```python
# creating a copy of the dataframe
df1 = df.copy(deep=True)
df1.head()
```

```python
# identify where df1 is a series or data frame
type(df1)
```

```python
# changing all columns in lower case
df1.columns = df1.columns.str.lower()
df1
```

```python
# checking missing values 
df1.isna().sum()
```

```python
df1.columns
```

```python
# checking all unique values in the column country
df1['country'].unique()
```

```python
# dropping a all other countries except United state in the column countries
df1.drop(df1[df1['country'] != 'United States'].index, inplace=True)
df1
```

```python
# confirming that united states is the only one left
df1['country'].unique()
```

```python
# dropping the longitudes and latitudes columns
df1 = df1.drop(['latitude', 'longitude','event.id'], axis=1)
```

```python
df1.describe()
```

```python
# filling categorical data using a variable
y_columns = ['injury.severity', 'aircraft.damage', 'total.fatal.injuries']

for col in y_columns:
    df1[col] = df1[col].fillna('Unknown')
df1
```

```python
# Drop columns with more than 50% missing values - schedule and air.carrier
threshold = 0.5 * len(df1)
df1 = df1.dropna(thresh=threshold, axis=1)
```

```python
# filling numerical columns total.fatal.injuries,number.of.engines using median

# Identify numerical columns
numerical_cols = ['total.fatal.injuries', 'number.of.engines']

# Fill missing values in each numerical column with its median
for col in numerical_cols:
    df1.loc[:, col] = pd.to_numeric(df1[col], errors='coerce') 
    df1[col] = df1[col].fillna(df1[col].median())
```

```python
# checking the number of missing values in the dataset
df1.isna().sum()
```

```python
# Only select numeric columns 
numeric_columns = df.select_dtypes(include=['number']).columns[:]

# Define grd layout 
rows, cols = 2,3
fig, axes = plt.subplots(rows, cols, figsize=(21,13))

# Flatten the arrays 
axes = axes.flatten()

# Plot histograms for numeric columns 
for i, column in enumerate(numeric_columns):
    sns.histplot(df[column], bins=50, ax=axes[i])
    axes[i].set_title(f'Histogram for {column}')
    axes[i].set_xlabel(column)
    axes[i].set_ylabel('Frequency')
    
    
# Hide empty subplots 
for j in range (i + 1, rows * cols):
    axes[j].axis('off')

plt.tight_layout()
plt.show()
```

Univariate Analysis

```python
# analysis purposes of flights

# Create the bar chart
flight = df1['purpose.of.flight'].value_counts().reset_index()
flight.columns = ['purpose_of_flight', 'count']

plt.figure(figsize=(10, 6))  # Increase figure size for clarity

# Create the bar chart
plt.bar(flight['purpose_of_flight'], flight['count'], color='skyblue')

# Add labels and title
plt.xlabel('Purpose of Flight', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.title('Flight Purpose Distribution', fontsize=16)

# Rotate x-axis labels to prevent overlap
plt.xticks(rotation=45, ha='right', fontsize=12)


# Add extra padding at the bottom for x-axis labels
plt.subplots_adjust(bottom=0.2)

# Display the plot
plt.tight_layout()
plt.show()
```

```python
 # Total fatal injuries by aircraft model
fatal_injuries_by_model = df1.groupby('model')['total.fatal.injuries'].sum().sort_values()

# Plot the top 10 safest models (least fatal injuries)
fatal_injuries_by_model.head(10).plot(kind='bar', title='Top 10 Safest Aircraft Models')
```

```python
# Count of accidents by flight phase
accidents_by_phase = df1['broad.phase.of.flight'].value_counts()

# Plot the flight phases with accidents
accidents_by_phase.plot(kind='bar', title='Accidents by Flight Phase')
```

```python
# Weather conditions and severity
weather_severity = df1.groupby('weather.condition')['total.fatal.injuries'].sum()

# Plot weather conditions vs. fatal injuries
weather_severity.plot(kind='bar', title='Weather Impact on Fatal Injuries')
```

```python
# Group by 'make' and 'model' and aggregate injuries
safety_by_make_model = df1.groupby(['make', 'model'])[
    ['total.fatal.injuries', 'total.serious.injuries', 'total.minor.injuries']
 ].sum()

# Add a 'total injuries' column
safety_by_make_model['total.injuries'] = safety_by_make_model.sum(axis=1)

# Get the safest aircraft (least total injuries)
safest_aircraft = safety_by_make_model.sort_values(by='total.injuries').head(10)
print(safest_aircraft)
```

```python
# Count accidents by engine type
accidents_by_engine = df1['engine.type'].value_counts()

# Total injuries by engine type
injuries_by_engine = df1.groupby('engine.type')['total.fatal.injuries'].sum()

# Plot injuries by engine type
injuries_by_engine.plot(kind='bar', title='Injuries by Engine Type')
```
